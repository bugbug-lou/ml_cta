```python
import sys
import os
import argparse
import time
import pandas as pd
import numpy as np
import seaborn as sns

sns.set(rc={'figure.figsize':(20,8)})

import warnings
if not sys.warnoptions:
    warnings.simplefilter("ignore")

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
```

```python
year = 2019
ticker = 'IFM'

# cd = os.path.dirname(__file__)
df = pd.read_csv(os.path.join(f'data/index/{year}/{ticker}.csv'), index_col=0, encoding='gbk')
df['type'] = df.index
df['time'] = pd.to_datetime(df['时间'])
df['day'] = df.time.dt.day
df['month'] = df.time.dt.month
df['open'] = df['开']
df['close'] = df['收']
df['high'] = df['高']
df['low'] = df['低']
df['volume'] = df['成交量']
df = df.set_index('time')
df['time'] = df.index
df = df[['time', 'month', 'day', 'open', 'close', 'high', 'low', 'volume']]
```

```python
df['v'] = df['close']  - df['close'].shift(1)
```

```python
df.v
```

```python
# df['mv2'] = (df.volume) *df.v*(np.abs(df.v))
df['mv2'] = df.volume *(np.abs(df.v))

# df['mv2'] = df.v*np.abs(df.v)
```

```python
df['energy'] = df.mv2.ewm(halflife=3).mean()
df['energy_long'] = df.mv2.abs().ewm(halflife=360).mean()
df['er'] = df['energy']/df['energy_long']
```

```python
day = 17
month = 4
df1 = df[(df.day==day) & (df.month==month)]
df1 = df1.iloc[10:,:]
```

```python
df1['er'].plot()
df1['close'].plot(secondary_y=True)

sns.despine()
```
